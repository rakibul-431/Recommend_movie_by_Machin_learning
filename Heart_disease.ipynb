```python
from numpy._core import numeric
target_col='HeartDisease'
numeric_col=['Age','RestingBP','Cholesterol','FastingBS','MaxHR','Oldpeak']
categorical_col=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']
```

```python
df['Age'].hist(bins=30,figsize=(8,4))
plt.title('Histogram of numerical of age',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
df[numeric_col].boxplot()
plt.title("it's depict the box plot")
plt.tight_layout()
plt.show()
```

```python
corre_matrix=df[numeric_col+[target_col]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corre_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title("Condition Matrix")
plt.tight_layout()
plt.show()
```
